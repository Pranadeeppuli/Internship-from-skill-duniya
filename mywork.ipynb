```python
#data pre-processsing

from sklearn.preprocessing import LabelEncoder
```

```python
#innistialising encoder
le=LabelEncoder()
```

```python
data['Gender']=le.fit_transform(data['Gender'])
data['Vehicle_Damage']=le.fit_transform(data['Vehicle_Damage'])
data['Vehicle_Age']=le.fit_transform(data['Vehicle_Age'])
```

```python
data.head()
```

```python
#separeting independend and depended variables
y=data['Response']
x=data.drop(['Vintage','id','Region_Code','Response'],axis=1)
```

```python
#scaling independent variables
from sklearn.preprocessing import StandardScaler
```

```python
names=x.columns
names
```

```python
#initialising scaler
scale = StandardScaler()
```

```python
x=scale.fit_transform(x)
x
```

```python
x=pd.DataFrame(x,columns=names)
```

```python
###training and testing splitting
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#model building
from sklearn.linear_model import LinearRegression
```

```python
#initialisig the model
model=LinearRegression()
model
```

```python
#fit/train the model
model.fit(x_train,y_train)
```

```python
#predicton
pred=model.predict(x_test)
```
